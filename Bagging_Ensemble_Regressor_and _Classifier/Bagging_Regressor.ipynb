```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
print("feature Names :" ,str(housing.feature_names))
print("Target Names :" ,str(housing.target_names))
print("Feature size :", str(housing.data.shape))
print("Target size :", str(housing.target.shape))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
```

```python
X_housing = housing.data
y_housing = housing.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_housing, y_housing, test_size=0.2, random_state=42)
```

```python
lr=LinearRegression()
DTR=DecisionTreeRegressor()
KNR=KNeighborsRegressor()
DTR.fit(X_train, y_train)
lr.fit(X_train, y_train)
KNR.fit(X_train, y_train)
```

```python
lr_pred = lr.predict(X_test)
DTR_pred = DTR.predict(X_test)
KNR_pred = KNR.predict(X_test)
```

```python
r2_score_lr = r2_score(y_test,lr_pred)
r2_score_DTR = r2_score(y_test,DTR_pred)
r2_score_KNR = r2_score(y_test,KNR_pred)

print("r2score of linear regression", r2_score_lr)
print("r2score of decision tree regressor",r2_score_DTR)
print("r2score of KNeighborsRegressor",r2_score_KNR)
```

```python
from sklearn.ensemble import BaggingRegressor
```

```python
bag_regressor = BaggingRegressor(random_state=1)
bag_regressor.fit(X_train, y_train)
```

```python
bag_regressor_pred = bag_regressor.predict(X_test)
r2_score_bag_regressor = r2_score(y_test,bag_regressor_pred)
print("r2score of bagging regressor",r2_score_bag_regressor)
```
